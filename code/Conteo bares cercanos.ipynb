## Read Bares y Restaurantes

```python
bares
```

### Calcular Bares y Restaurantes cercanos en 500 metros

```python
def conteo_en_radio(cais_SDA,establecimientos_df,radio,nombre_var):
    
    for cai in cais_SDA.CAIIEPOLIC :
        cai_lat_lon = cais_SDA[cais_SDA.CAIIEPOLIC==cai]['lat_lon']
        temp_df = establecimientos_df.copy()
        temp_df["distancia"] = temp_df.lat_lon.apply(lambda x: distance.distance(cai_lat_lon, x).km)
        cais_SDA.loc[cais_SDA.CAIIEPOLIC==cai,nombre_var]=len(temp_df.sort_values("distancia").query("distancia<"+str(radio)).lat_lon)

    return cais_SDA
```

```python
cais_SDA = conteo_en_radio(cais_SDA,bares,0.5,"conteo_bares_500m")
cais_SDA = conteo_en_radio(cais_SDA,bares,1,"conteo_bares_1km")
cais_SDA = conteo_en_radio(cais_SDA,bares,2,"conteo_bares_2km")
```

## Read Manzanas

```python
manzanas_df = pd.read_excel("../data/otros/manzanas/MGN_ANM_MANZANA.xlsx")
```

```python
manzanas_df.head()
```

```python
list(manzanas_df.columns)
```

```python
not_variable = [
    'DPTO_CCDGO', 'MPIO_CCDGO', 'MPIO_CDPMP', 'CLAS_CCDGO', 'SETR_CCDGO',
    'SETR_CCNCT', 'SECR_CCDGO', 'SECR_CCNCT', 'ZU_CCDGO', 'ZU_CDIVI',
    'SETU_CCDGO', 'SETU_CCNCT', 'SECU_CCDGO', 'SECU_CCNCT', 'MANZ_CCDGO',
    'AG_CCDGO', 'DATO_ANM', 'VERSION', 'Shape_Leng', 'DDLat', 'DDLon',
    'ORIG_OID', 'NMB_LC_CM', 'TP_LC_CM '
]
```

```python
manzanas_var = [x for x in list(manzanas_df.columns) if x not in not_variable]
```

```python
manzanas = manzanas_df[manzanas_var].copy()
```

```python
manzanas["lat_lon"] = list(zip(manzanas.LATITUD, manzanas.LONGITUD))
```

```python
manzanas_resultados = pd.DataFrame()
radio = 0.5
for cai in cais_SDA.CAIIEPOLIC:
    cai_lat_lon = cais_SDA[cais_SDA.CAIIEPOLIC==cai]['lat_lon']
    temp_df = manzanas.copy()
    temp_df["distancia"] = temp_df.lat_lon.apply(lambda x: distance.distance(cai_lat_lon, x).km)
    suma_manzanas = temp_df.sort_values("distancia").query("distancia<"+str(radio)).sum()
    suma_manzanas['CAIIEPOLIC']=cai
    manzanas_resultados = manzanas_resultados.append(suma_manzanas,ignore_index=True)
```

```python
manzanas_resultados.head()
```

```python
manzanas_resultados.drop(['CD_LC_CM','COD_DANE_A','TP_LC_CM','distancia','lat_lon','LATITUD','LONGITUD'],axis=1,inplace=True)
```

## Final Merge

```python
final_df = cais_SDA.merge(manzanas_resultados,how='left',on='CAIIEPOLIC')
```

```python
final_df.head()
```

```python
final_df.to_csv('../data/interim/nuevas_variables.csv',index=False)
```

## Read SDA Data

```python
SDA_data = pd.read_csv(interim_path + 'SDA_data.csv', dtype='unicode', encoding='utf_8')
```

```python
SDA_data.shape
```

```python
SDA_data.head()
```

```python
SDA_data.drop(['Ponderación','Tipo'],axis=1,inplace=True)
```

```python
SDA_data.columns= ["Estación","Serial","Variable","Fecha","Valor","Ponderación","Tipo"]
```

```python
# 1. Remove special character when loading the source file
SDA_data['Estación'] = SDA_data['Estación'].str.replace('�','')
SDA_data['Estación'] = SDA_data['Estación'].str.replace('"','')
# due to problems with enconding, some additional relacement needs to occur:
SDA_data['Estación'] = SDA_data['Estación'].str.replace(u'\ufeff','')
# Due to the replacement on (1), a word got problems, so we need to take care of these values:
SDA_data['Estación'] = SDA_data['Estación'].str.replace('Fontibn','Fontibón')
```

```python
SDA_data.Valor=SDA_data.Valor.astype('float')
```

```python
SDA_data.head(5)
```

### Median Leq

```python
SDA_mean = SDA_data[SDA_data["Variable"]=="Leq"][["Estación","Valor"]].groupby("Estación").agg("median").reset_index(drop=False)
```

```python
SDA_mean.head()
```

### Median Leq per hour

```python
SDA_data["Hora"] = pd.to_datetime(SDA_data.Fecha).dt.hour
```

```python
SDA_hour_mean = SDA_data[SDA_data["Variable"]=="Leq"][["Estación","Hora","Valor"]].groupby(["Estación","Hora"]).agg("median").reset_index(drop=False)
```
