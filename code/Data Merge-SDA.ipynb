```python
import pandas as pd
import numpy as np
import glob 
import datetime
```

```python
def load_file_to_df(search_path,df_size=20):
    files = glob.glob(search_path,recursive = True)
    print('Number of files to process: ',len(files))
    df = pd.DataFrame()
    counter =0
    
    for filepath in files:
        file = open(filepath, 'r')
        print (filepath,': ',len(file.readlines())) 
        file.close()
        df_aux = pd.read_csv(filepath,delimiter =';')
        df_aux.drop(['Unidad','Ponderación'],axis=1,inplace=True)
        df = df.append(df_aux)
        counter += 1
        
        if counter%df_size==0 or counter==len(files):
            df.to_csv(output_path+str(counter/df_size)+'SDA_data.csv',index=False)
            df = pd.DataFrame()
            print('Dataframe saved'+str(round(counter/df_size,1)))
            
    print('Total number of records:',df.shape[0])
    print('Total number of columns:',df.shape[1])
    return df
```

```python
def load_file_to_pivot(search_path):
    files = glob.glob(search_path,recursive = True)
    print('Number of files to process: ',len(files))
    df = pd.DataFrame()
    
    for filepath in files:
        file = open(filepath, 'r')
        print (filepath,': ',len(file.readlines())) 
        file.close()
        df_aux = pd.read_csv(filepath,delimiter =',')
        df_aux["Tipo"].fillna('Normal',inplace=True)
        df_aux_wide = df_aux.pivot_table(index=['Estación', 'Serial', 'Fecha'],columns=['Componente','Variable','Tipo'],values=['Valor'])
        df_aux_wide.columns = df_aux_wide.columns.droplevel()
        df_aux_wide.columns = ["_".join(x) for x in df_aux_wide .columns.ravel()]
        df_aux_wide.reset_index(inplace=True)
        #df_aux_wide.Estación = df_aux_wide.Estación.apply(lambda x: x.replace('"', ""))
        df = df.append(df_aux)

    return df
```
